##Handling Missing Values

```python
df.isna().sum()
```

```python
df['Zipcode'] = df['Zipcode'].fillna(df['Zipcode'].mean())
```

```python
df['Latitude'] = df['Latitude'].fillna(df['Latitude'].mean())
```

```python
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].mean())
```

```python
df['Area of the House from Basement (in Sqft)'] = df['Area of the House from Basement (in Sqft)'].fillna(df['Area of the House from Basement (in Sqft)'].median())
```

```python
df['Living Area after Renovation (in Sqft)'] = df['Living Area after Renovation (in Sqft)'].fillna(df['Living Area after Renovation (in Sqft)'].median())
```

```python
df['No of Bathrooms'] = df['No of Bathrooms'].fillna(df['No of Bathrooms'].median())
```

```python
df['No of Times Visited'] = df['No of Times Visited'].fillna(df['No of Times Visited'].mode()[0])
```

```python
df['Sale Price'] = df['Sale Price'].fillna(df['Sale Price'].mean())
```

```python
df['Flat Area (in Sqft)'] = df['Flat Area (in Sqft)'].fillna(df['Flat Area (in Sqft)'].mean())
```

```python
df['Lot Area (in Sqft)'] = df['Lot Area (in Sqft)'].fillna(df['Lot Area (in Sqft)'].mean())
```

```python
original_rows = df.shape[0]

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()

    print(f"{col} : removed {outliers_count} outliers")

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print(f"\nTotal rows before removal: {original_rows}")
print(f"Total rows after removal: {df.shape[0]}")
print(f"Total rows removed overall: {original_rows - df.shape[0]}")
```
