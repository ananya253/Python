```python
import pandas as pd
import requests
```

###Step 1: Load SpaceX Launch Data from API

```python
launch_url = 'https://api.spacexdata.com/v4/launches'
launch_response = requests.get(launch_url)
launch_response
```

```python
launch_data = launch_response.json()
df_launch = pd.DataFrame(launch_data)
launches=df_launch[['name','date_utc','success','details','rocket']]
launches
```

```python
launches['date_utc'] = pd.to_datetime(launches['date_utc'])
launches['year'] = launches['date_utc'].dt.year
launches.head()
```

###Step 2: Load Rocket Metadata

```python
rocket_url = 'https://api.spacexdata.com/v4/rockets'
rocket_response = requests.get(rocket_url)
rocket_response
```

```python
rocket_data = rocket_response.json()
df_rocket = pd.DataFrame(rocket_data)
rockets = df_rocket[['id','name','type','active','stages']]
rockets
```

###Step 3: Merge Launch and Rocket Data


```python
merged_df = pd.merge(launches, rockets, left_on='rocket', right_on='id')
merged_df = merged_df.drop(['rocket'],axis=1)
merged_df
```

```python
import random
```

###Step 4: Add Simulated Country Information

```python
countries=['USA','Russia','India','China','France']
merged_df['country']=[random.choice(countries) for i in range(len(merged_df))]
merged_df
```

###Step 5: Store Merged Data in SQLite3

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('launches.db')
merged_df.to_sql('launches', conn, if_exists='replace', index=False)
```

```python
pd.read_sql_query('SELECT * FROM launches LIMIT 5;', conn)
```

###Step 6: Run SQL Queries on the Data to analyze


1. Launches by Country


```python
query1 = """SELECT country, COUNT(*) AS launch_count
FROM launches
GROUP BY country
ORDER BY launch_count DESC;"""

pd.read_sql_query(query1, conn)
```

2. Which year had the highest number of launches?

```python
query2 = """SELECT year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC
LIMIT 1;"""

pd.read_sql_query(query2, conn)
```

3. Top 5 Missions by Launch Count

```python
query3 = """SELECT name_x AS mission_name, COUNT(*) AS launch_count
FROM launches
GROUP BY name_x
ORDER BY launch_count DESC
LIMIT 5;"""

pd.read_sql_query(query3, conn)
```
